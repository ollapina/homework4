## Задание 1

Дан список вида:

```python
data = [
    [13, 25, 23, 34],
    [45, 32, 44, 47],
    [12, 33, 23, 95],
    [13, 53, 34, 35]
]
```

Напишите функцию, которая возвращает сумму элементов на диагонали. Т. е. 13+32+23+35.  
Список может быть любой длины, но всегда является "квадратным" (количество элементов во вложенных списках равно их количеству).

```python
def sum_elements(data):
    sum_list = []
    x = 0
    i = 0
    for e in data:
        if x < len(e):
            sum_list.append(e[i])
            x += 1
            i += 1
#     print(sum_list)
    print(sum(sum_list))
```

```python
sum_elements(data)
```

```python
def sum_el(data):
    return sum(value[i] for i, value in enumerate(data))
```

```python
sum_el(data)
```

## Задание 2

Напишите функцию, возвращающую сумму первых n чисел Фибоначчи (для этого вам может понадобится ознакомиться с рекурсией).

```python
def fibonacci(n):
    if n < 2:
        return 1
    else:
        res = fibonacci(n-1) + fibonacci(n-2)
        return res
```

```python
def fibonacci_sum():
    n = int(input('Введите число: '))
    fibonacci(n)
    summa = 0
    for i in range(0, n):
        r = fibonacci(i)
        summa += r
#         print(r)
    print(summa)
```

```python
fibonacci_sum()
```

```python
one_list = ['2018-01-01', 'yandex', 'cpc', 100]
```

```python
def create_dict(one_list):
    matryoshka_dict = one_list[-1]
    for i in one_list[-2::-1]:
        matryoshka_dict = {i:matryoshka_dict}
    print(matryoshka_dict)
```

## Задание 4

Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
def check_command(command, commands_dict):
    for key in commands_dict.keys():
        if command == key:
            return True
    return False
```

```python
def run_commands():
    commands_dict = dict(p = 'people', s = 'shelf', l = 'listing', ads = 'add_shelf', d = 'delete', m = 'move', a = 'add')
    print(commands_dict)
    command = input('Введите команду :')
    print(check_command(command, commands_dict))
    print(commands_dict[command])
    if commands_dict[command] == 'people':
        search_people()
    elif commands_dict[command] == 'shelf':
        search_shelf()
    elif commands_dict[command] == 'listing':
        request_listing_docs()
    elif commands_dict[command] == 'add_shelf':
        adding_shelf()
    elif commands_dict[command] == 'delete':
        remove_doc_from_list_and_shelf()
    elif commands_dict[command] == 'move':
        move_doc_on_shelf()
    elif commands_dict[command] == 'add':
        add_doc()
```

```python
run_commands()
print(documents)
print(directories)
```

```python
def search_people():
    """
    функция для исполнения команды, которая спросит номер документа
    и выведет имя человека, которому он принадлежит
    """
    doc_number = input('Введите номер документа: ')
    for doc in documents:
        if doc_number == doc['number']:
            return doc['name']
    return 'Указанный документ отсутствует в каталоге!'
```

```python
search_people()
```

```python
def search_shelf():
    """
    функция для исполнения команды, которая спросит номер документа
    и выведет номер полки, на которой он находится
    """
    doc_number = input('Введите номер документа: ')
    for key, value in directories.items():
        if doc_number in value:
            return key
    return 'Указанный документ отсутствует в каталоге!'
```

```python
def request_listing_docs():
    """
    функция для исполнения команды, которая выведет список всех документов в формате
    passport "2207 876234" "Василий Гупкин"
    """
    for doc in documents:
        print(doc['type'], '"'+ doc['number'] + '"', '"' + doc['name'] + '"')
```

```python
def adding_shelf():
    """
    функция для исполнения команды, которая спросит номер новой полки и добавит ее в перечень
    """
    new_shelf_number = input('Введите номер новой полки: ')
    if new_shelf_number not in directories:
        directories.update({new_shelf_number:[]})
    return 'Указанная полка уже существует!'
```

```python
def remove_doc_from_list_and_shelf():
    """
    функция для исполнения команды, которая спросит номер документа
    и удалит его из каталога и из перечня полок
    """
    doc_number = input('Введите номер документа: ')
    for doc in documents:
        if doc_number == doc['number']:
            documents.remove(doc)
    for value in directories.values():
        if doc_number in value:
            value.remove(doc_number)
    return 'Указанный документ отсутствует в каталоге!'
```

```python
def move_doc_on_shelf():
    """
    функция для исполнения команды, которая спросит номер документа
    и целевую полку и переместит его с текущей полки на целевую
    """
    doc_number = input('Введите номер документа: ')
    for doc in documents:
        if doc_number == doc['number']:
            shelf_number = input('Введите номер полки: ')
            if shelf_number in directories:
                for value in directories.values():
                    if doc_number in value:
                        value.remove(doc_number)
                        directories[shelf_number].append(doc_number)
                        return
            return 'Указанная полка не существует!'
    return 'Указанный документ отсутствует в каталоге!'
```

```python
def add_doc():
    """
    функция для исполнения команды, которая добавит новый документ в каталог
    и в перечень полок, спросив его номер, тип, имя владельца
    и номер полки, на котором он будет храниться
    """
    shelf_number = input('Введите номер полки: ')
    if shelf_number not in directories:
        print('Указанная полка не существует!')
        return
    doc_number = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    doc_name = input('Введите имя и фамилию владельца документа: ')
    documents.append({'type': doc_type, 'number': doc_number, 'name': doc_name})
    directories[shelf_number].append(doc_number)
    return
```
